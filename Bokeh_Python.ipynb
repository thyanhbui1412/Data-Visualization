# Median Debt by Region by Gender

```python
#compiling source 
data1 = {'region' : Region,
        'y1'   : Female,
        'y2'   : MedianDebt,
        'y3': Male}
source = ColumnDataSource(data=data1)
```

```python
# instantiating the figure object 

graph = figure(x_range=Region, y_range=(0, 60000), title="Median Debt by Region by Gender",width=900,
           height=750, toolbar_location=None, tools="")

#graph
for size,y,colors,titl in zip([0.20,0,-0.20],['y1','y2','y3'],Spectral4,["Female","MedianDebt", "Male"]):
    graph.vbar(x=dodge('region', size, range=graph.x_range), top=y, source=source,width=0.2, color=colors, legend_label=titl)


#extra formats
graph.add_layout(Title(text="New England (CT, ME, MA, NH, RI, VT)\nMid East (DE, DC, MD, NJ, NY, PA)\nGreat Lakes (IL, IN, MI, OH, WI)\nPlains (IA, KS, MN, MO, NE, ND, SD)\nSoutheast (AL, AR, FL, GA, KY, LA, MS, NC, SC, TN, VA, WV)\nSouthwest (AZ, NM, OK, TX)\nRocky Mountains (CO, ID, MT, UT, WY)\nFar West (AK, CA, HI, NV, OR, WA)\n*Median debt = The median original amount of the loan principal upon entering repayment", align="left"), "below")
graph.y_range.start = 0
graph.x_range.range_padding = 0.1
graph.xaxis.major_label_orientation = 1
graph.xgrid.grid_line_color = None
graph.legend.location = "top_right"
graph.legend.click_policy="hide"
graph.y_range.start = 5000
graph.y_range.end = 18000

show(graph)
```

# Median Debt by Region by Gender

```python
data = {'region' : Region,
        'y1'   : MaleEarning,
        'y2'   : NonMaleEarning}

source = ColumnDataSource(data=data)
p = figure(x_range=Region, y_range=(0, 60000), title="Median Earning By Region By Gender",width=900,
           height=450, toolbar_location=None, tools="")

for size,y,colors,titl in zip([0.15,-0.15],['y1','y2'],Spectral4,["MaleEarning","NonMaleEarning"]):
    p.vbar(x=dodge('region', size, range=p.x_range), top=y, source=source,width=0.2, color=colors, legend_label=titl)

p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "horizontal"
p.add_layout(Title(text="New England (CT, ME, MA, NH, RI, VT)\nMid East (DE, DC, MD, NJ, NY, PA)\nGreat Lakes (IL, IN, MI, OH, WI)\nPlains (IA, KS, MN, MO, NE, ND, SD)\nSoutheast (AL, AR, FL, GA, KY, LA, MS, NC, SC, TN, VA, WV)\nSouthwest (AZ, NM, OK, TX)\nRocky Mountains (CO, ID, MT, UT, WY)\nFar West (AK, CA, HI, NV, OR, WA)\n", align="left"), "below")

show(p)
```
